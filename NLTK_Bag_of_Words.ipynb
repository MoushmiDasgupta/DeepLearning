```python
filelink = '/content/drive/MyDrive/Colab Notebooks/NLP/book.txt'
```

```python
file = open(filelink,'r')
```

```python
text = file.read()
```

```python
text
```

```python
text.replace('\ufeffOne','One')
```

```python
tokens = text.split()
tokens
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
newtokens = [w for w in tokens if not w in stop_words]
newtokens
```

```python
newtokens = [w for w in tokens if w.isalpha]
newtokens
```

```python
newtokens = [w for w in tokens if len(w) >2]
newtokens
```

```python
import re
import string
```

```python
newpuntuations = re.compile('[%s]'% re.escape(string.punctuation))
```

```python
newtokens = [newpuntuations.sub('',w) for w in newtokens]
```
